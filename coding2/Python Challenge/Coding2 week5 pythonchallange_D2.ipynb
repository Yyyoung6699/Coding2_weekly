```python
import urllib.request
import re

html = urllib.request.urlopen("http://www.pythonchallenge.com/pc/def/ocr.html").read().decode()
comments = re.findall("<!--(.*?)-->", html, re.DOTALL)[-1]

# Create a dictionary that holds the frequency of each character
freq = {}


for c in comments:
    # Just think about letters
    if c.isalpha():
        if c in freq:
            freq[c] += 1
        else:
            freq[c] = 1

# Output each character and its frequency
for c, count in freq.items():
    print(c, count)
```
